```python
def water_jug_problem():
    try:
        jug1_capacity = int(input("Jug1 capacity: "))
        jug2_capacity = int(input("Jug2 capacity: "))
        target = int(input("Target amount: "))
    except ValueError:
        print("Error: Please enter valid integer values for capacities and target.")
        return

    if jug1_capacity <= 0 or jug2_capacity <= 0 or target <= 0:
        print("Error: Capacities and target must be positive numbers.")
        return
    if target > jug1_capacity and target > jug2_capacity:
        print("Error: Target cannot be greater than both jug capacities.")
        return

    jug1 = 0
    jug2 = 0

    while True:
        print(f"\nCurrent state -> Jug1: {jug1}L, Jug2: {jug2}L")
        print("Choose an action:")
        print("1. Fill Jug1")
        print("2. Fill Jug2")
        print("3. Empty Jug1")
        print("4. Empty Jug2")
        print("5. Pour Jug1 into Jug2")
        print("6. Pour Jug2 into Jug1")
        print("7. Exit")

        try:
            choice = int(input("Your choice: "))
        except ValueError:
            print("Error: Please enter a number from 1 to 7.")
            continue

        match choice:
            case 1:
                jug1 = jug1_capacity
            case 2:
                jug2 = jug2_capacity
            case 3:
                jug1 = 0
            case 4:
                jug2 = 0
            case 5:
                transfer = min(jug1, jug2_capacity - jug2)
                jug1 -= transfer
                jug2 += transfer
            case 6:
                transfer = min(jug2, jug1_capacity - jug1)
                jug2 -= transfer
                jug1 += transfer
            case 7:
                print("Exiting.")
                break
            case _:
                print("Invalid choice. Try again.")

        if jug1 == target or jug2 == target:
            print(f"Target achieved! Jug1: {jug1}L, Jug2: {jug2}L")
            break

water_jug_problem()
```

```python
def solve_water_jug(x_cap, y_cap, target):
    x = 0
    y = 0
    step = 0

    while True:
        print(f"Step {step}: Jug X = {x}/{x_cap}, Jug Y = {y}/{y_cap}")

        if x == target or y == target:
            print("Target reached!")
            break

        if x == 0:
            x = x_cap
        elif y != y_cap:
            pour_amount = min(x, y_cap - y)
            x -= pour_amount
            y += pour_amount
        else:
            y = 0

        step += 1

        if step > 100:
            print("No solution found within reasonable steps.")
            break

solve_water_jug(4, 3, 2)
```

# **Missionaries and Cannibals**

```python
def is_valid(m_left, c_left, m_right, c_right):
    if (m_left > 0 and c_left > m_left) or (m_right > 0 and c_right > m_right):
        return False
    if min(m_left, c_left, m_right, c_right) < 0 or max(m_left, c_left, m_right, c_right) > 3:
        return False
    return True

def main():
    m_left, c_left = 3, 3
    m_right, c_right = 0, 0
    boat = 'left'
    step = 0

    print("Initial state:")
    print(f"Left bank: M={m_left}, C={c_left} | Right bank: M={m_right}, C={c_right} | Boat: {boat}")

    while not (m_left == 0 and c_left == 0):
        step += 1
        print(f"\nStep {step}:")
        print(f"Current state - Left bank: M={m_left}, C={c_left} | Right bank: M={m_right}, C={c_right} | Boat: {boat}")
        try:
            move_m = int(input("Number of missionaries to move: "))
            move_c = int(input("Number of cannibals to move: "))
        except ValueError:
            print("Please enter valid integers.")
            continue

        if move_m + move_c == 0 or move_m + move_c > 2:
            print("Invalid move: You must move 1 or 2 people.")
            continue

        if boat == 'left':
            if move_m > m_left or move_c > c_left:
                print("Invalid move: Not enough people on the left bank.")
                continue
            m_left -= move_m
            c_left -= move_c
            m_right += move_m
            c_right += move_c
            boat = 'right'
        else:
            if move_m > m_right or move_c > c_right:
                print("Invalid move: Not enough people on the right bank.")
                continue
            m_right -= move_m
            c_right -= move_c
            m_left += move_m
            c_left += move_c
            boat = 'left'

        if not is_valid(m_left, c_left, m_right, c_right):
            print(" Missionaries will be eaten  by cannibals.")
            break

    if m_left == 0 and c_left == 0:
        print("All missionaries and cannibals have crossed safely.")

if __name__ == "__main__":
  main()
```
